```python
# pip install numpy pandas matplotlib sklearn tensorflow -i https://pypi.doubanio.com/simple
import tensorflow as tf
print('Tensorflow Version: {}'.format(tf.__version__))
```

```python
import pandas as pd
print('Pandas Version: {}'.format(pd.__version__))
data = pd.read_csv('./dataset/Income.csv')
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(data.Education, data.Income)
```

```python
x = data.Education
y = data.Income
```

```python
# 创建模型
model = tf.keras.Sequential() # 顺序模型
# Dense y = ax + b
model.add(tf.keras.layers.Dense(1, input_shape=(1,))) 
model.summary()
```

```python
# 训练
model.compile(optimizer='adam', loss='mse') # 指定优化器 和 减少误差算法(mse平方差)
history = model.fit(x, y, epochs=6000) # 填充训练数据x变量和y结果， 重复训练6000次
```

```python
model.predict(x)
```

```python
# 用数据测试训练结果，输入Education 20， 会自动计算学习来的Income
model.predict(pd.Series([20]))
```
