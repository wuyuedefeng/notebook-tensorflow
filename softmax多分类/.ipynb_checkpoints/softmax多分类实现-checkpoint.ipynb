```python
import tensorflow as tf
print('Tensorflow Version: {}'.format(tf.__version__))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(train_image, train_label), (test_image, test_label) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
train_image.shape
```

```python
train_label.shape
```

```python
test_image.shape
```

```python
# 查看第一张图片真实存储数据结构
# train_image[0]
# 预览第一张图片
plt.imshow(train_image[0])
```

```python
train_label # 这里是用数字标注物品的种类
```

```python
# 如果用独热编码的label， loss使用categorical_crossentropy
train_label_onehot = tf.keras.utils.to_categorical(train_label)
train_label_onehot
```

```python
train_image = train_image/255 # 先将像素点的值缩减到0-1,再进行分类
test_image = test_image/255 # 将像素点的值缩减到0-1
```

```python
train_image.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28,28))) # 展平成28*28的向量
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
# metrics: acc -> 准确度
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
model.fit(train_image, train_label, epochs=5)
```

```python
# 用测试数据进行评价
model.evaluate(test_image, test_label)
```

```python
# 重新创建模型演示 过拟合
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28,28))) # 展平成28*28的向量
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
# 添加 validation_data，可以打印训练模型相对于测试数据的val_loss和val_acc
history = model.fit(train_image, train_label, epochs=5, validation_data=(test_image, test_label))
```

```python
# 损失值对比
plt.plot(history.epoch, history.history.get('loss'), label='loss')
plt.plot(history.epoch, history.history.get('val_loss'), label='val_loss')
plt.legend()
```

```python
# 正确率对比
plt.plot(history.epoch, history.history.get('acc'), label='acc')
plt.plot(history.epoch, history.history.get('val_acc'), label='val_acc')
plt.legend()
```

过拟合：在训练数据上得分很高，在测试数据上得分相对比较低（最好的解决办法是增加训练数据， 其他办法 -> 1. dropout 人为丢弃一些层 2. 正则化  3. 图像增强）

欠拟合：在训练数据上得分比较低，在测试数据上得分相对比较低 -> 

```python
# 添加dropout解决过拟合
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28,28))) # 展平成28*28的向量
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dropout(0.5)) # 随机丢弃50%
model.add(tf.keras.layers.Dense(10, activation='softmax'))
# 编译 训练
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
# 添加 validation_data，可以打印训练模型相对于测试数据的val_loss和val_acc
history = model.fit(train_image, train_label, epochs=5, validation_data=(test_image, test_label))
# 损失值对比
plt.plot(history.epoch, history.history.get('loss'), label='loss')
plt.plot(history.epoch, history.history.get('val_loss'), label='val_loss')
plt.legend()
```

```python
# 正确率对比
plt.plot(history.epoch, history.history.get('acc'), label='acc')
plt.plot(history.epoch, history.history.get('val_acc'), label='val_acc')
plt.legend()
```
